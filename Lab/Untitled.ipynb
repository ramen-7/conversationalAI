```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_data(pageNumber):
    url = "https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_" + str(pageNumber) + "?ie=UTF8&pg=" + str(pageNumber)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')#r.content gives bytecode, r.text gives unicode
    complete = []
    for d in soup.find_all('div', attrs={'class':'zg-grid-general-faceout'}):
        book_name = d.find('div', attrs={'class': '_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y'})
        author_name = d.find('div', attrs={'class': 'a-row a-size-small'})
        ratings = d.find('span', attrs={'class': 'a-icon-alt'})
        link = d.find('a', attrs={'class': 'a-link-normal'})
        sub = []
        
        if book_name is not None:
            sub.append(book_name.text)
        else:
            sub.append('NaN')
            
        if author_name is not None:
            sub.append(author_name.text)
        else:
            sub.append('NaN')
            
        if ratings is not None:
            sub.append(ratings.text)
        else:
            sub.append('NaN')
            
        if link is not None:
            sub.append('https://www.amazon.in' + link.get('href'))
        else:
            sub.append('NaN')
        
        complete.append(sub)
        
    return complete
```

```python
get_data(1)
```

```python
number_of_pages = 2
result = []
for i in range(1, number_of_pages+1):
    result.append(get_data(i))
```

```python
list2 = []
def flatten(list1):
    for sublist1 in list1:
        for sublist2 in sublist1:
            list2.append(sublist2)
    return list2
```

```python
import pandas as pd
df = pd.DataFrame(flatten(result), columns=['Book Name', 'Author Name', 'Ratings', 'Link'])
```
